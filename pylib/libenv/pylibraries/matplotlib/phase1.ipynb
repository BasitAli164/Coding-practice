```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
```

# Line Chart or Line Graph

```python
months=[1,2,3,4]
sales=[1111,1000,3000,1400]
```

```python
plt.plot(months,sales,color="green",linestyle="--",linewidth=1,label="2025 Sales Data of University",marker="o")
plt.xlabel("Moths") # show the label of x-axies
plt.ylabel("Sales") # show the label of y-axies
plt.title("2025 Sales Data with Line Chart") # show the title of the graph
plt.legend(loc="upper left",fontsize=8) # show the label inside the plot function 
plt.grid(color="red",linewidth=1,linestyle=':') # show the square box like graph paper
plt.xlim(1,4) # give limits means 1,4 becuase i have data of 4 month
plt.ylim(0,4000) # here also give limits the sales
plt.xticks([1,2,3,4],['M1','M2','M3','M4']) # here we replace the 
plt.show() # show the graph
```

# Bar Chart

```python
product=['A','B','C','D']
sale=[1000,1440,500,3000]
plt.bar(product,sale,color='brown',label='2025 Sales Data') # if you required horizontally then use .barh
plt.legend() 
plt.title("2025 Product Sales Data with Bar Chart")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.grid()
```

# Pie Chart

```python
region=['north','south','east','west']
revenue=[3000,2400,5000,1000]
plt.pie(revenue,labels=region, autopct='1.1f%%',colors=['gold','green','pink','orange'])
plt.title("Revenue By Region")
plt.show()
```
