```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
```

```python
months=[1,2,3,4]
sales=[1111,1000,30,4000]
```

```python
plt.plot(months,sales,color="green",linestyle="--",linewidth=1,label="2025 Sales Data of University")
plt.xlabel("Moths") # show the label of x-axies
plt.ylabel("Sales") # show the label of y-axies
plt.title("2025 Sales Data") # show the title of the graph
plt.legend(loc="upper left",fontsize=8) # show the label inside the plot function 
plt.grid() # show the square box like graph paper
plt.show() # show the graph
```
