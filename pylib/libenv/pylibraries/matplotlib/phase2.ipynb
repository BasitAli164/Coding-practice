# Scatter Plots
#### Most use this to find the relationship between tow variables

```python
import matplotlib.pyplot as plt
```

```python
hours_Stdy=[1,2,3,4,5,6,7,8]
marks=[34,45,55,67,70,75,85,90]
plt.scatter(hours_Stdy,marks,color="green",marker='^',label="Student Data")
plt.xlabel("Hours Study")
plt.ylabel("Exam Scores")
plt.title("Relationship between study time & Scores")
plt.legend()
plt.grid()
plt.show()
```

```python
hours_Stdy=[1,2,3,4,5,6,7,8]
marks=[34,45,55,67,70,75,85,90]

hoursStd=[1,2,3,4,5,6,7,8]
scores=[40,45,50,60,65,70,75,80]
plt.scatter(hours_Stdy,marks,color="green",marker='^',label="Student A Data")
plt.scatter(hoursStd,scores,color="red",marker='o',label="Student B Data")
plt.xlabel("Hours Study")
plt.ylabel("Exam Scores")
plt.title("Comparsion of Different Class Std")
plt.legend()
plt.grid()
plt.show()
```

# Sub Plots

```python


fig,ax=plt.subplots(1,2,figsize=(10,5)) # 1st row, 2nd column , 1st position/ index
x=[1,2,3,4]
y=[10,20,15,25]


ax[0].plot(x,y,color="red")
ax[0].set_title("Line Plot")


ax[1].bar(x,y)
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of Line and Bar Chart") # main Title


plt.tight_layout() # use for adjustment

plt.show()
```

# How to save the graph

```python
plt.plot(x,y,color="green",marker="o")
plt.title('Line Plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.savefig("linePlot.png",dpi=300,bbox_inches='tight')
plt.show()
```
