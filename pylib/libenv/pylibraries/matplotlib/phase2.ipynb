# Scatter Plots
#### Most use this to find the relationship between tow variables

```python
import matplotlib.pyplot as plt
```

```python
hours_Stdy=[1,2,3,4,5,6,7,8]
marks=[34,45,55,67,70,75,85,90]
plt.scatter(hours_Stdy,marks,color="green",marker='o',label="Student Data")
plt.xlabel("Hours Study")
plt.ylabel("Exam Scores")
plt.title("Relationship between study time & Scores")
plt.legend()
plt.grid()
plt.show()
```
